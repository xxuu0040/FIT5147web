```python
df = data[['ACCIDENT_DATE','ACCIDENT_TIME','ACCIDENT_TYPE','DAY_OF_WEEK','LGA_NAME','LIGHT_CONDITION','ROAD_GEOMETRY','TOTAL_PERSONS',
 'INJ_OR_FATAL','FATALITY','SERIOUSINJURY','OTHERINJURY','NONINJURED','MALES','FEMALES','BICYCLIST','PASSENGER','DRIVER','PEDESTRIAN','PILLION',
 'MOTORIST', 'OLD_DRIVER','YOUNG_DRIVER','RMA','DCA_CODE','SPEED_ZONE','SEVERITY']]
df
```

```python
df['year'] = pd.DatetimeIndex(df['ACCIDENT_DATE']).year
df['month'] = df.apply(lambda x: x['ACCIDENT_DATE'].split('/')[1],axis=1)
df['hour'] = df.apply(lambda x: x['ACCIDENT_TIME'].split('.')[0],axis=1)
df
```

```python
df_new = pd.read_csv("output.csv", index_col=None, header=0)
df_new['DAY_OF_WEEK'].unique()
```

```python
df_new['LGA_NAME'].replace('SHEPPARTON','GREATER SHEPPARTON',inplace=True)
df_new['LGA_NAME'].unique()
```

```python
df_new.to_csv("output.csv",index=False)
```

```python
df_new['DAY_OF_WEEK'].replace('0', change(df_new), inplace=True)
df_new['DAY_OF_WEEK'].unique()
```

```python
df_new['DAY_OF_WEEK'] = df_new['DAY_OF_WEEK'].fillna(method='ffill')
df_new['DAY_OF_WEEK'].unique()
```

```python
def change(df):
    for i in range(len(df['DAY_OF_WEEK'])):
        if df['DAY_OF_WEEK'][i] == '0':
            df['DAY_OF_WEEK'][i] = str(pd.DatetimeIndex(df['date'][i],freq='D').day_name)
```

```python
df_new['date'] = df_new.apply(lambda x: x['ACCIDENT_DATE'].split('/')[2] + '-' + x['ACCIDENT_DATE'].split('/')[1] + '-' + x['ACCIDENT_DATE'].split('/')[0], axis=1)
df_new
```

```python
df_new['hour'] =df_new.apply(lambda x: x['ACCIDENT_TIME'].split('.')[0],axis=1)
df_new['hour'].unique()
```

```python
df_new['DAY_OF_WEEK'].unique()
```

```python
df = df_new.groupby(['LGA_NAME','year']).agg({'index': ['sum'],'INJ_OR_FATAL': ['sum']})
df.columns = ['Accident', 'Injury']
df = df.reset_index()
df
```
